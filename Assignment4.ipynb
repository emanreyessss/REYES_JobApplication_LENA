```python
import random
import pandas as pd
import names
```

Function for creating Team Table

```python
def gen_team(n):
    team_names = set()
    while len(team_names) < n:
        team_names.add('Team ' + names.get_last_name())
    teams = pd.DataFrame({'Team_id': list(range(1, n + 1)),'Team_Name': list(team_names)})
    return teams
```

Function for creating Matches Table

```python
def gen_match(team, number_of_matches):
    uniq_match = (len(team) - 1) * len(team) // 2

    #Raise Error due to the Input
    if number_of_matches < uniq_match:
      raise ValueError(f"Input Number of Matches ({number_of_matches}) is less than the possible Unique Matches ({uniq_match}).\n"
      + f"Input Greater than {uniq_match}")

    #Generate Unique Matches
    match_once = [(team.iloc[i, 0], team.iloc[l, 0]) for i in range(len(team)) for l in range(i + 1, len(team))]

    # Adding matches that may repeat
    add_match = []
    while len(add_match) < number_of_matches - uniq_match:
        x, y = random.sample(range(1,len(team)), 2)  # Generate two unique indices
        if x!=y:
          add_match.append((x, y))

    # Combine the unique and additional matches
    matches = match_once + list(add_match)

    # Randomized row of the matches and Tuple of W/L
    random.shuffle(matches)
    wl_shuffle = [(y, x) if random.choice([True, False]) else (x, y) for x, y in matches]

    # Prepare DataFrame
    matches_df = pd.DataFrame({
        'MATCH_NUMBER': range(1, number_of_matches + 1),
        'WINNER': [x for x, y in wl_shuffle],
        'LOSER': [y for x, y in wl_shuffle]
    })

    return matches_df
```

```python
teams = gen_team(20)
teams
```

```python
matches = gen_match(teams,200)
matches
```
